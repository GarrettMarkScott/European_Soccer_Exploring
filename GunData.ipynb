Exploring the states with the most handguns, long_guns, and other in 2016 to match population data.

```python
#Filter to data from 2016
guns2016 = df2[df2['month'].str.contains('2016')]
guns2016
```

The code below find the number of guns checks for the year of 2016 segmented by state

```python
guncounts = df2[df2['month'].str.contains('2016')].groupby('state')['handgun','long_gun','other'].sum().sort_index().astype(int)
guncounts['total_gun_checks'] = guncounts[['handgun','long_gun','other']].sum(axis=1)
guncounts
```

Now time to explore the data for population to mix with state to find gun owners/capita. Below is a list of options

```python
state_population = df1.T.iloc[2:,[0]].sort_index()
state_population.columns = ['state_population']
state_population
```

Now we want to merge the gun data with the population data

```python
merged_data = guncounts.join(state_population, how='inner')
merged_data
```

```python
type(merged_data['state_population'][2])
```

```python
merged_data.state_population = merged_data.state_population.str.replace(',','').astype(int)
```

```python
type(merged_data['state_population'][2])
```

```python
merged_data['percent_checked'] = merged_data['total_gun_checks'] / merged_data['state_population']
merged_data.sort_values('percent_checked', ascending=False)
```
